# Top 5 cities by Sales,Profit,Discount

```python
total_sales = data_set.groupby('City')['Sales'].sum()

top_5_cities = total_sales.sort_values(ascending = False).iloc[0:5]
top_5_cities.plot(kind = 'barh')
plt.title('Top 5 cities by sales',fontsize=15)
plt.xticks(rotation=270)
plt.xlabel('Sales')
plt.show()
```

```python
total_profit = data_set.groupby('City')['Profit'].sum()

top_5_cities = total_profit.sort_values(ascending= False).iloc[0:5]
top_5_cities.plot(kind = 'barh')
plt.title('Top 5 cities by Profit',fontsize=15)
plt.ylabel('Profit')
```

```python
total_profit = data_set.groupby('City')['Discount'].sum()

top_5_cities = total_profit.sort_values(ascending= False).iloc[0:5]
top_5_cities.plot(kind = 'bar')
plt.title('Top 5 cities by Discount',fontsize=15)
plt.xticks(rotation=0)
plt.ylabel('Discount')
```

# Sales by Region

```python
total_sales = data_set.groupby('Region')['Sales'].sum()

region_sales = total_sales.sort_values(ascending = False).iloc[0:4]
region_sales.plot(kind ='barh')
plt.title('Sales by Region',fontsize=15)
plt.xlabel('Sales')
plt.show()
```

# Sales by Segment

```python
total_profit = data_set.groupby('Segment')['Sales'].sum()

segment_sales = total_profit.sort_values(ascending= False).iloc[0:4]
segment_sales.plot(kind = 'bar',fontsize=10)
plt.title('Sales by Segment')
plt.xticks(rotation=0)
plt.ylabel('Sales')
```

# Average profit by Discount

```python
ax = sns.barplot(x = 'Discount', y = 'Profit', data = data_set, palette = 'cubehelix_r', ci = None )
ax.set_title('Average profit by Discount', fontsize = 18)
plt.axhline(y = data_set['Profit'].mean(), linestyle='--')
plt.axhline(y = 0, linestyle='-')
plt.xlabel("")
plt.ylabel("Average profit")
plt.show()
```

# Average Sales by Discount

```python
ax = sns.barplot(x = 'Discount', y = 'Sales', data = data_set, palette = 'gist_stern_r', ci = None )
ax.set_title('Average Sales by Discount', fontsize = 18)
plt.axhline(y = data_set['Sales'].mean(),linestyle='--')
plt.axhline(y = 0, linestyle='-')
plt.xlabel("")
plt.ylabel("Average Sales")
plt.show()
```

**From above we can see that, sales made with less than 20% discounts are more profitable.**

# Average Sales and profit/loss for Region

```python
df_region = data_set.groupby(['Region'])[['Sales','Profit']].mean()
df_region.sort_values('Profit', ascending = False)[['Sales', 'Profit']].plot(kind = 'bar',figsize = (15, 5),color=['teal','red'])
plt.ylabel('Average Profit/Loss and Sales')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.title('Average Sales and Profit/Loss for Regions', fontsize = 15)
plt.show()
```

```python
df_region = data_set.groupby(['Category'])[['Sales','Profit']].mean()
df_region.sort_values('Profit', ascending = False)[['Sales', 'Profit']].plot(kind = 'bar',figsize = (15, 5),color=['green','red'])
plt.ylabel('Average Profit/Loss and Sales')
plt.xlabel('Region')
plt.xticks(rotation=25)
plt.title('Average Sales and Profit/Loss for Category', fontsize = 10)
plt.show()
```
